```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(26,6))
df_bar = df.groupby('species')['sepal_length'].mean().reset_index(name='count')
df_bar
df_bar.plot(kind='bar',x='species',y='count')
```

```python
sns.barplot(df,x='species',y='sepal_length')
```

```python
plt.bar(df_bar['species'], df_bar['count'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count by Species')
plt.show()
```

```python
px.bar(df_bar,x='species',y='count')
```
