```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
```

```python
df.plot(kind='line',x='species',y='sepal_length',marker='o')
```

```python
sns.lineplot(df,x='species',y='sepal_width',marker='o')
```

```python
px.line(df,x='species',y='sepal_width')
```

```python
px.line_3d(df,x='species',y='sepal_width',z='sepal_length')
```

```python
px.line(df,x='sepal_length',y='sepal_width',color='species')
```
