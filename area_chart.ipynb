```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
```

```python
df_area = df.groupby('species')['sepal_length'].mean().reset_index(name='count')
df_area 
```

```python
df_area.plot(kind='area',x='species',y='count')
```

```python
plt.plot(df_area['species'], df_area['count'], color="Slateblue", alpha=0.6)
```

```python
px.area(df_area,x='species',y='count')
```
