```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
```

```python
df_pie = df.groupby('species')['sepal_length'].mean().reset_index(name='count')
df_pie.plot(kind='pie',x='species',y='count',autopct='%1.1f%%', startangle=140)
```

```python
plt.pie(df_pie['count'],labels=df_pie['species'],autopct='%1.1f%%', startangle=140)
```

```python
px.pie(df_pie,names='species',values='count')
```
