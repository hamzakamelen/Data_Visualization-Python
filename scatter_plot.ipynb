```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.title('Scatter Plot - Relation b/w Sepal length and width')
```

```python
sns.scatterplot(df,x='sepal_length',y='sepal_width',hue='species')
plt.title('Scatter Plot - Relation b/w Sepal length and width')
```

```python
plt.scatter(data=df,x='sepal_length',y='sepal_width')
```

```python
px.scatter(df,x='sepal_length',y='sepal_width',title='Scatter Plot - Relation b/w Sepal length and width')
```

```python
px.scatter(df,x='sepal_length' ,y='sepal_width',color='species')
```

```python
px.scatter(df,x='sepal_length' ,y='sepal_width',size='petal_length',color='species')
```

```python
px.scatter_3d(df,x='sepal_width',y='sepal_length',z='petal_width',color='species')
```

```python
px.scatter_ternary(df, a='sepal_length', b='sepal_width', c='petal_width', color='species')
```
